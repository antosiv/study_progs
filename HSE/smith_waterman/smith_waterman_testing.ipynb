```python
from smith_waterman import align
```

### simple test

```python
_ = align('TGTTACGG', 'GGTTGACTA', similar_score=1, gap_penalty=1, mismatch_penalty=1)
```

### random sequence with gaps

```python
seq = ''.join(np.random.choice(['T', 'C', 'G', 'A'], 15))
seq_with_gaps = ''.join(item if np.random.uniform() > 0.3 else '-' for item in seq)
print(seq, seq_with_gaps, sep='\n')
```

```python
_ = align(seq, ''.join(filter(lambda x: x!= '-', seq_with_gaps)))
```

### two different random sequences

```python
seq1 = ''.join(np.random.choice(['T', 'C', 'G', 'A'], 12))
seq2 = ''.join(np.random.choice(['T', 'C', 'G', 'A'], 12))
print(seq1, seq2, sep='\n')
```

```python
_ = align(seq1, seq2)
```
