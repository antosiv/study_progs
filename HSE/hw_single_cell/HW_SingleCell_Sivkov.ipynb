```r
data.processed = NormalizeData(cbmc)
data.processed = FindVariableFeatures(data.processed, selection.method="vst", nfeatures=2000)
data.processed = ScaleData(data.processed, verbose = FALSE)
data.processed = RunPCA(data.processed, npcs = 30, verbose = FALSE)
data.processed = RunUMAP(data.processed, reduction = "pca", dims = 1:20)
data.processed = FindNeighbors(data.processed, reduction = "pca", dims = 1:20)
data.processed = FindClusters(data.processed, resolution = 0.5)
```

Кластеризация отличается, некоторые кластеры из аннотации поделены в несколько, некоторые наоборот не разделены, но по большей части аннотация и полученная кластеризация выглядят обоснованными и похожими

```r
options(repr.plot.width=20, repr.plot.height=10)
plot.1 <- DimPlot(data.processed, reduction = "umap", group.by = "rna_annotations")
plot.2 <- DimPlot(data.processed, reduction = "umap", label = TRUE)
plot_grid(plot.1, plot.2)
```

```r
data.markers.cluster = FindAllMarkers(data.processed, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25)
```

```r
top10 = data.markers.cluster %>% group_by(cluster) %>% top_n(n = 10, wt = avg_logFC)
DoHeatmap(object = data.processed, features = top10$gene, label=TRUE) + NoLegend()
```

```r
top1 = data.markers.cluster %>% group_by(cluster) %>% top_n(n = 1, wt = avg_logFC)
FeaturePlot(data.processed, features = top1$gene)
```

```r
Idents(data.processed) = data.processed$rna_annotations
data.markers.cell = FindAllMarkers(
    data.processed, only.pos = TRUE, min.pct = 0.25, logfc.threshold = 0.25
)
```

```r
top10 = data.markers.cell %>% group_by(cluster) %>% top_n(n = 10, wt = avg_logFC)
DoHeatmap(object = data.processed, features = top10$gene, label=TRUE) + NoLegend()
```

```r
top1 = data.markers.cell %>% group_by(cluster) %>% top_n(n = 1, wt = avg_logFC)
FeaturePlot(data.processed, features = top1$gene)
```

```r
gene.markers = getMarkerGenes(
  object = data.processed,
  organism = 'hg',
  groups = c('seurat_clusters'),
  name = 'cerebro_seurat',
  only_pos = TRUE,
  min_pct = 0.7,
  thresh_logFC = 0.25,
  thresh_p_val = 0.01,
  test = 'wilcox',
  verbose = TRUE
)

enrichment.data = getEnrichedPathways(
  object = gene.markers,
  databases = c("GO_Cellular_Component_2018")
)
```

```r
enriment.result = attr(enrichment.data, 'misc')$enriched_pathways$cerebro_seurat_enrichr$seurat_clusters
```

```r
enriment.result[enriment.result['seurat_clusters'] == 6,]
```

В 6 кластер попал процесс T cell receptor complex, это можно объяснить тем что в этот кластер входят T-клетки (судя по аннотации); Также там встречается ещё процесс, связанный с иммунной системой (phagocytic vesicle membrane). Всё остальное связано с housekeeping-генами и рибосомами, возможно если отфильтровать такие процессы, то можно найти ещё специфичных для кластера процессов.

Также отбросили образцы с > 30% рибосомной РНК. Даллее был использован Seurat, выделены наиболее вариабельные гены, запущен UMAP, выполнена кластеризация (по сути, повторён анализ с семинара).
